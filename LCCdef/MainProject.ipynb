```python

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=8)
a=[[0,0]]  

plt.clf() 
plt.gcf().clear()
xm=0
ym=1
for i in range(width1):
    for j in range(height1):
        coordinate = x,y=i,j 
        points1=image1.getpixel(coordinate)
        points2=image2.getpixel(coordinate)
        points3=image3.getpixel(coordinate)
        points4=image4.getpixel(coordinate)
        dif1=points1[xm] - points1[ym]
        dif2=points2[xm] - points2[ym]
        dif3=points3[xm] - points3[ym]
        dif4=points1[xm] - points4[ym]
        dif1=abs(dif1)
        dif2=abs(dif2)
        dif3=abs(dif3)
        dif4=abs(dif4)
        if dif1 > chk :
            if points1[xm] < max_skip and points1[ym] < max_skip:   
                a=np.insert(a,1,[points1[xm],points1[ym]],axis=0)
        if dif2 > chk :
            if points2[xm] < max_skip and points2[ym] < max_skip:   
                a=np.insert(a,1,[points2[xm],points2[ym]],axis=0)
        if dif3 > chk :
            if points3[xm] < max_skip and points3[ym] < max_skip:   
                a=np.insert(a,1,[points3[xm],points3[ym]],axis=0)
        if dif4 > chk :
            if points4[xm] < max_skip and points4[ym] < max_skip:   
                a=np.insert(a,1,[points4[xm],points4[ym]],axis=0)        

                
kmeans.fit(a)
centroids=kmeans.cluster_centers_
labels=kmeans.labels_


for n in range(len(a)):
    plt.plot(a[n][0],a[n][1],colors[labels[n]])
    
#plt.scatter(centroids[:,0],centroids[:,1],marker="x",s=700)
plt.show()
```
